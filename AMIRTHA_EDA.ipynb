```python
from sklearn.linear_model import LinearRegression,Lasso,Ridge,ElasticNet
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
df_cleaned.loc[:,'Time_of_Day']=df_cleaned['Time_of_Day'].map(Time_of_day_map)
df_cleaned.loc[:,'Traffic_Conditions']=df_cleaned['Traffic_Conditions'].map(traffic_map)
df_cleaned.loc[:,'Weather']=df_cleaned['Weather'].map(weather_map)
```

```python
data = pd.concat([x, Y], axis=1).dropna()
x, Y = data.iloc[:, :-1], data.iloc[:, -1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,Y_train,Y_test=train_test_split(x,Y,test_size=0.60,random_state=21)
```

```python
regression=LinearRegression()
regression.fit(x_train,Y_train) 
```

```python
import numpy as np
def evaluate_model(true, predicted):
    mae = mean_absolute_error(true, predicted)
    mse = mean_squared_error(true, predicted)
    r2_square = r2_score(true, predicted)
    return mae, r2_square,mse
```

```python
models={
    'LinearRegression':LinearRegression(),
    'Lasso':Lasso(),
    'Ridge':Ridge(),
    'Elasticnet':ElasticNet(),
}

for i in range (len(list((models)))):
    print(i)
    
```

```python
models={
    'LinearRegression':LinearRegression(),
    'Lasso':Lasso(),
    'Ridge':Ridge(),
    'Elasticnet':ElasticNet(),
}
trained_model_list=[]
model_list=[]
r2_list=[]

for i in range(len(list(models))):
    model=list(models.values())[i]
    model.fit(x_train,Y_train)

    #Make Predictions
    y_pred=model.predict(x_test)

    mae, r2_square,mse=evaluate_model(Y_test,y_pred)

    print(list(models.keys())[i])
    model_list.append(list(models.keys())[i])
    

    print('Model Training Performance')
   
    print("MAE:",mae)
    print("R2 score",r2_square)

    r2_list.append(r2_square)
    
    print('='*35)
    print('\n')
```
